# What is Auto ML?

Auto ML can be understood as automatic machine learning.Let's understand it in terms of the data scientist activities.Let's suppose that we have a classification problem to solve.Then the classical approach in Supervised Machine Learning for solving the problem consists of:
1. A knowledge of a programming language like R or Python or any other language.
2. We should  have a machine with  all the necessary dependecies installed.
3. we'd load the data into a predefined data structure and try to figure out which features to use.
4. Data preprocessing for cleaning and imputating the data.
5. writing some lines of code to split that data into training and test sets.
6. Writing some lines of code to select a classifier and then to do a training based on the training dataset.
7. After training, evaluating the model against the test data and then doing a parameter tuning to get the maximum accuracy.

But what if you are someone who doesn't know a programming language? What if there is something which saves you from all this hastle?
What if obtaining Machine Learning results is as easy as click of few buttons?

That's what Auto ML does. It makes obtaining the machine learning results a pretty easy task, just a click of few buttons and you are done.

# TransmogrifAI
* TransmorgrifAI is AutoML library which runs on the top of ApacheSpark.
* It is and end to end automated machine learning library for data structures.


# TransmogrifAI work flow
* Normally research and development that typically goes in building a good machine learning model but the data preparation, Feature engineering and training of model is an iterative process. 
* It takes time more than weks or even months.
* TransmogrifAI is the librabry that helps the data scentists to automate the data cleansing, feature engineering and model selection to achieve the perfect model.

* TransmorgifAI consists of five components.
They are:
        1.Feature Inference
        2.Transmogrification
        3.Automated Feature Validation
        4.Automated Model Selection
        5.Hyperparameter Optimization
<b>1.Feature Inference :</b>
* The first step in Machine learning(ML) is data preparation.
* In this datascientist will collect all the data from resourses and make them flatten,joins,aggregates and populated in to the flexible data structure commonly known as data frame.
* This data frames will use in the pipelines and the errors will occurs due to multiplication of strings occur(for example).
* This is the main drawback, so the transmorgif provides us the time saving and it can done by itself of collecting the raw data and response signals as features.
* Features are strongly typed, and TransmogrifAI supports a rich and extensible feature type hierarchy.
* This hierarchy will supports different types of codes like zip codes,geo_location,phone numbers and etc.

<b> 2.Transmogrification:</b>
* The data is transformed in to the numerical representation and it can be easily exploited by the machine learning algorithms.This process is known as feature engineering.
* There are many ways to trasnfrom the data but doing in the right way is the art of data science.
* For example if we have features like city name We can transform the features in to the numerical values but if we have thousands of features then its not possible to transfrom and dont know the correct way of doing.
* Automatic engineering of dozens of different feature types into numeric vectors is what gives TransmogrifAI its name. TransmogrifAI comes with a myriad of techniques for all the supported feature types ranging from phone numbers, email addresses, geo-locations to text data even.

<b>3.Automated feature validation:</b>
* Transmogrification have the algorithms that can perform automatic feature validation to remove features.
* These algorithms are useful when we are dealing with high dimensionality data and unknon data.They apply statistical based tests based on features.

<b>4.Automated model selection:</b>
* The final stage of datascientists process is to choose the model that will fit best to predict.choosing correct algorithm and parameters is the tough task.
* In Automated model selection TransmogrifAI will apply seveeral algorithms on data and use the average validation error to choose the best algorithm.
* It also deals with the unbalanced data by appropriate sampling the data.

<b>5.Hyperparameter Optimization:</b>
* Hyperparameter Optimixation is like tunable knobs on the machine learning algorithms.
* The tuning of parameters is the time taking process.it may take some weeks to months but with the help of TransmogrifAI will reduce to some hours by adding some lines of code.
